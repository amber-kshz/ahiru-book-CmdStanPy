```python
import numpy as np
import pandas as pd
from cmdstanpy import CmdStanModel
```

```python
df = pd.read_csv("../chap08/data/data-salary-2.txt")
print(len(df))
df.head(5)
```

```python
data = {
    "N": len(df),
    "K": len(set(df["KID"].values)),
    "X": df["X"].values,
    "Y": df["Y"].values,
    "KID": df["KID"].values
}
```

# Model 10-5

```python
model10_5 = CmdStanModel(stan_file="model/model10-5.stan")
print(model10_5)
```

```python
fit10_5 = model10_5.sample(data=data, show_progress=True)
```

```python
fit10_5.summary()
```

`R_hat`の値から、収束していないことがわかる。

# Model 10-6

```python
model10_6 = CmdStanModel(stan_file="model/model10-6.stan")
print(model10_6)
```

```python
fit10_6 = model10_6.sample(data=data, show_progress=True)
```

```python
summary10_6 = fit10_6.summary()
summary10_6[summary10_6["R_hat"] >= 1.1]
```

```python
summary10_6
```
