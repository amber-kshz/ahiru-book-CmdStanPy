```python
import pandas as pd
from cmdstanpy import CmdStanModel
```

```python
df = pd.read_csv("../chap08/data/data-salary-2.txt")
print(len(df))
df.head(5)
```

```python
data = {
    "N": len(df),
    "K": 4,
    "X": df["X"].values,
    "Y": df["Y"].values,
    "KID": df["KID"].values
}
```

# Model 8-4b

```python
model8_4b = CmdStanModel(stan_file="../chap09/model/model8-4b.stan")
print(model8_4b)
```

```python
fit8_4b = model8_4b.sample(data=data)
```

```python
fit8_4b.summary()
```

# Model 8-4c

```python
model8_4c = CmdStanModel(stan_file="model/model8-4c.stan")
print(model8_4c)
```
