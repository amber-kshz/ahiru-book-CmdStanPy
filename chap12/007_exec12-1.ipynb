```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def gen_time_series(T, s_mu, s_Y, mu0):
    mu = np.zeros(T)
    Y = np.zeros(T)
    mu[0] = mu0
    Y[0] = mu[0] + np.random.randn() + s_Y
    for t in range(1, T):
        mu[t] = mu[t-1] + np.random.randn() * s_mu
        Y[t] = mu[t] + np.random.randn() + s_Y
    return mu, Y
```

```python
def show_time_series(ax, T, s_mu, s_Y, mu0):
    mu, Y = gen_time_series(T, s_mu, s_Y, mu0)
    ax.plot(mu, '.-', label="mu")
    ax.plot(Y, '.-', label="Y")
    ax.legend()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6), sharex=True, sharey=True)

list_s_mu = [0.1, 1.0]
list_s_Y = [1.0, 0.1]

for ax, s_mu, s_Y in zip(axes, list_s_mu, list_s_Y):
    show_time_series(ax=ax, T=50, s_mu=s_mu, s_Y=s_Y, mu0=0.0)
    ax.set_title(f"s_mu = {s_mu}, s_Y = {s_Y}")

fig.show()
```
