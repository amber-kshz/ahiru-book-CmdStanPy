```python
import pandas as pd
from cmdstanpy import CmdStanModel
```

```python
df = pd.read_csv("../chapter08/data/data-salary-2.txt")
print(len(df))
df.head(5)
```

```python
data = {
    "N": len(df),
    "K": 4,
    "X": df["X"].values,
    "Y": df["Y"].values,
    "KID": df["KID"].values
}
```

```python
model = CmdStanModel(stan_file="model/exercise11-1.stan")
print(model)
```

```python
fit = model.sample(data=data)
```

```python
summary = fit.summary()
summary
```
